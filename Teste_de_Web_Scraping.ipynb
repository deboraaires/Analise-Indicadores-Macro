```
import requests
from bs4 import BeautifulSoup as bs

#  pegando todo o conteúdo de um requisição get na url do Clima tempo
html = requests.get("https://www.climatempo.com.br/previsao-do-tempo/cidade/558/saopaulo-sp").content

# é criado um objeto chamado soup que está interpretando o documento HTML.
soup = bs(html, 'html.parser')

# podemos ver que ele retornará todo o HTML da página
print(soup.prettify())
```

```
temperatura = soup.find("span", class_="-gray-light")

print(temperatura.string)
```

```
URL = 'https://www.gov.br/anp/pt-br/centrais-de-conteudo/publicacoes/boletins-anp/boletim-mensal-da-producao-de-petroleo-e-gas-natural'
FILETYPE = '.xlsm'

def get_soup(url):
  return bs(requests.get(url).text, 'html.parser')

for link in get_soup(URL).find_all('a', class_="internal-link"):
  file_link = link.get('href')
  if FILETYPE in file_link:
    print(file_link)

    # The with statement simplifies exception handling by encapsulating common preparation and cleanup tasks.
    # Opening a file using with is as simple as: with open(filename) as file:
    # Use file to refer to the file object
    with open(link.text + FILETYPE,'wb') as file: 
      response = requests.get(file_link)
      file.write(response.content)
      break
```
